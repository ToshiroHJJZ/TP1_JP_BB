```python
white=df['white'].mean()
df['white'].fillna(white,inplace=True)
```

```python
black=df['black'].mean()
df['black'].fillna(black,inplace=True)
```

```python
hispanic=df['hispanic'].mean()
df['hispanic'].fillna(hispanic,inplace=True)
```

```python
asian=df['asian'].mean()
df['asian'].fillna(asian,inplace=True)
```

```python
american_indian_alaska_native=df['american_indian_alaska_native'].mean()
df['american_indian_alaska_native'].fillna(american_indian_alaska_native,inplace=True)
```

```python
hawaiian_native_pacific_islander=df['hawaiian_native_pacific_islander'].mean()
df['hawaiian_native_pacific_islander'].fillna(hawaiian_native_pacific_islander,inplace=True)
```

```python
df['weapon_source'].fillna('unknown',inplace=True)
```

```python
df['weapon'].fillna('unknown',inplace=True)
```

```python
low_grade=df['low_grade'].mode()
df['low_grade'].fillna(low_grade,inplace=True)
```

```python
high_grade=df['high_grade'].mode()
df['high_grade'].fillna(high_grade,inplace=True)
```

```python
df['time'].fillna('12:00 PM',inplace=True)
```

```python
# checking that all values we imputed for have a null sum of 0
df.isnull().sum()
```

```python
df['hispanic']=df['hispanic'].str.replace(',', '', regex=True).astype(float)
```

```python
casualties_by_day = df.groupby('day_of_week')['casualties'].sum().reset_index()
plt.bar(casualties_by_day['day_of_week'], casualties_by_day['casualties'])

# Set labels and title
plt.xlabel('Day of the Week')
plt.ylabel('Casualties')
plt.title('Casualties by Day of the Week')

# Display the chart
plt.show()
```

```python
types=df.groupby('shooting_type')['killed'].sum().reset_index()
```

```python

filtered_types = types[types['shooting_type'].isin(['indiscriminate', 'targeted'])]
plt.bar(filtered_types['shooting_type'], filtered_types['killed'])
# Set labels and title
plt.xlabel('Shooting Type')
plt.ylabel('Killed')
plt.title('Deaths by Shooting Type')

# Display the chart
plt.show()
```

```python
df['low_grade'] = df['low_grade'].replace({'K': '0'})
```

```python
df['low_grade']=df['low_grade'].astype(int)
```

```python
df['high_grade'].value_counts()
```

```python
df['high_grade'] = df['high_grade'].replace({'PK': '0'})
```

```python
df['high_grade']=df['high_grade'].astype(int)
```

```python
mean_age_by_type = df.groupby('shooting_type')[['low_grade', 'high_grade']].mean()

# Create a bar chart to visualize the mean age
mean_age_by_type.plot(kind='bar')

# Set labels and title
plt.xlabel('Shooting Type')
plt.ylabel('Mean Age')
plt.title('Mean Age of Victims by Shooting Type')

# Display the chart
plt.show()
```

```python
df['enrollment'] = df['enrollment'].astype(int)

# Create the scatter plot
plt.figure(figsize=(10, 6))  # Set the size of the plot

sns.regplot(data=df, x='enrollment', y='casualties', line_kws={"color": "red"},ci=False)


# Set labels and title
plt.xlabel('Enrollment')
plt.ylabel('Casualties')
plt.title('Enrollment vs. Casualties in School Shootings')

# Display the plot
plt.show()
```
